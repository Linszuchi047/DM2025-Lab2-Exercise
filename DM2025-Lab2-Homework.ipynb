```python
### 1. Preprocessing Steps  (load + merge + clean text, with emoji & hashtags)

import json
import re
import pandas as pd

# 1.1 讀入 final_posts.json 並攤平成表格
posts = []
with open("dm-lab-2-private-competition/final_posts.json", "r", encoding="utf-8") as f:
    data = json.load(f)   # 檔案是一個 list，每個元素一篇貼文

for obj in data:
    post = obj["root"]["_source"]["post"]
    hashtags = post.get("hashtags") or []
    # 把 hashtag 變成有意義 token，保留情緒線索，例如 hashtag_sad, hashtag_happy
    hashtags_tokens = ["hashtag_" + h.lower() for h in hashtags]

    posts.append({
        "id": post["post_id"],                 # 對應 CSV 裡的 id
        "text": post.get("text", ""),
        "hashtags": " ".join(hashtags_tokens)
    })

df_posts = pd.DataFrame(posts)
print("posts shape:", df_posts.shape)

# 1.2 讀入標籤與 train/test split
df_emotion = pd.read_csv("dm-lab-2-private-competition/emotion.csv")              # columns: id, emotion
df_split   = pd.read_csv("dm-lab-2-private-competition/data_identification.csv")  # columns: id, split (train/test)

print(df_emotion.head())
print(df_split["split"].value_counts())

# 1.3 合併成一個 DataFrame
df = df_posts.merge(df_split, on="id", how="inner")
df = df.merge(df_emotion, on="id", how="left")  # test 沒有 emotion，會是 NaN

print(df.head())
print(df["split"].value_counts())

# 1.4 emoji / emoticon 映射表（先轉成英文 token 再做 regex）
EMOJI_MAP = {
    "😂": "emoji_laugh",
    "🤣": "emoji_laugh",
    "😢": "emoji_cry",
    "😭": "emoji_cry",
    "😡": "emoji_angry",
    "😠": "emoji_angry",
    "😔": "emoji_sad",
    "😞": "emoji_sad",
    "😕": "emoji_confused",
    "😨": "emoji_fear",
    "😱": "emoji_fear",
    "😍": "emoji_love",
    "❤": "emoji_love",
    "❤️": "emoji_love",
    "😊": "emoji_happy",
    "😁": "emoji_happy",
    "😃": "emoji_happy",
    "😄": "emoji_happy",
    "😆": "emoji_happy",
    "😅": "emoji_happy",
    "😒": "emoji_unamused",
    "😤": "emoji_frustrated",
    "👍": "emoji_thumbsup",
    "🙏": "emoji_pray",
    "💔": "emoji_brokenheart",
}

EMOTICON_MAP = {
    ":)": "smiley_happy",
    ":-)": "smiley_happy",
    ":D": "smiley_happy",
    ":-D": "smiley_happy",
    ":(": "smiley_sad",
    ":-(": "smiley_sad",
    ":'(": "smiley_cry",
    ":'-)": "smiley_cry",
    ":/')": "smiley_cry",
}

def replace_emojis_and_emoticons(text: str) -> str:
    if not isinstance(text, str):
        return ""
    # 先處理文字表情
    for emot, token in EMOTICON_MAP.items():
        text = text.replace(emot, f" {token} ")
    # 再處理 emoji
    for emo, token in EMOJI_MAP.items():
        text = text.replace(emo, f" {token} ")
    return text

# 1.5 定義文字清理函式
def clean_text(text: str) -> str:
    if not isinstance(text, str):
        return ""
    # 先保留 emoji/emoticon 對應 token
    text = replace_emojis_and_emoticons(text)
    text = text.lower()                                        # 全部小寫
    text = re.sub(r"http\S+|www\.\S+", " ", text)              # 移除網址
    # 只保留英文小寫、空白與底線（讓 hashtag_sad、emoji_laugh 這種詞保留下來）
    text = re.sub(r"[^a-z\s_]+", " ", text)
    text = re.sub(r"\s+", " ", text).strip()                   # 合併多重空白
    return text

# 把內容文字 + hashtags 合在一起做清理
df["clean_text"] = (df["text"].fillna("") + " " +
                    df["hashtags"].fillna("")).apply(clean_text)

print(df[["id", "clean_text"]].head())

# 1.6 切出 train / test DataFrame
train_df = df[df["split"] == "train"].copy()
test_df  = df[df["split"] == "test"].copy()

print("Train size:", len(train_df))
print("Test size :", len(test_df))
```

## 2. Feature Engineering Steps

```python
### Feature Engineering Steps  (TF-IDF + encode labels)

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder

# 2.1 建立 TF-IDF 向量器
tfidf = TfidfVectorizer(
    max_features=30000,   # 取前 30k 個常見詞
    ngram_range=(1, 2),   # unigram + bigram
    min_df=3              # 至少在 3 篇文中出現
)

# 2.2 將文字轉成稀疏矩陣
X_train = tfidf.fit_transform(train_df["clean_text"])
X_test  = tfidf.transform(test_df["clean_text"])

print("X_train shape:", X_train.shape)
print("X_test  shape:", X_test.shape)

# 2.3 將文字標籤 emotion 轉成數字
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(train_df["emotion"])

print("Classes:", label_encoder.classes_)
```

```python
### 2. Feature Engineering Steps  (word TF-IDF + char TF-IDF + label encode)

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from scipy.sparse import hstack

# 2.1 word-level TF-IDF
word_tfidf = TfidfVectorizer(
    max_features=40000,     # 取前 40k 個常見詞
    ngram_range=(1, 2),     # unigram + bigram
    min_df=3
)

# 2.2 char-level TF-IDF（可以抓到錯字、emoji 文字、lmao, omg 之類 pattern）
char_tfidf = TfidfVectorizer(
    analyzer="char",
    ngram_range=(3, 5),
    min_df=5,
    max_features=50000      # 限制字元特徵數，避免太大
)

X_train_word = word_tfidf.fit_transform(train_df["clean_text"])
X_test_word  = word_tfidf.transform(test_df["clean_text"])

X_train_char = char_tfidf.fit_transform(train_df["clean_text"])
X_test_char  = char_tfidf.transform(test_df["clean_text"])

# 2.3 合併兩種特徵
X_train = hstack([X_train_word, X_train_char])
X_test  = hstack([X_test_word, X_test_char])

print("X_train shape:", X_train.shape)
print("X_test  shape:", X_test.shape)

# 2.4 將文字標籤 emotion 轉成數字
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(train_df["emotion"])

print("Classes:", label_encoder.classes_)
```

```python
### 2. Feature Engineering Steps  (word TF-IDF + char TF-IDF + label encode, with sublinear_tf)

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from scipy.sparse import hstack

# 2.1 word-level TF-IDF
word_tfidf = TfidfVectorizer(
    max_features=50000,     # 可以視記憶體狀況再調整，例如 40000~60000
    ngram_range=(1, 2),     # unigram + bigram
    min_df=3,
    sublinear_tf=True,      # 使用 log(1 + tf) 對 SVM 通常較友善
)

# 2.2 char-level TF-IDF（可以抓到錯字、縮寫、口語 pattern）
char_tfidf = TfidfVectorizer(
    analyzer="char",
    ngram_range=(3, 5),
    min_df=5,
    max_features=60000,     # 限制字元特徵數，避免過大
    sublinear_tf=True,
)

X_train_word = word_tfidf.fit_transform(train_df["clean_text"])
X_test_word  = word_tfidf.transform(test_df["clean_text"])

X_train_char = char_tfidf.fit_transform(train_df["clean_text"])
X_test_char  = char_tfidf.transform(test_df["clean_text"])

# 2.3 合併兩種特徵
X_train = hstack([X_train_word, X_train_char])
X_test  = hstack([X_test_word, X_test_char])

print("X_train shape:", X_train.shape)
print("X_test  shape:", X_test.shape)

# 2.4 將文字標籤 emotion 轉成數字
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(train_df["emotion"])

print("Classes:", label_encoder.classes_)
```

## 3. Model Implementation Steps

```python
### Model Implementation Steps  (Linear SVM + validation + submission)

from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import f1_score
import pandas as pd

# 3.1 切一點 validation set 來看本地 F1（不影響最後訓練）
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train,
    test_size=0.2,
    random_state=42,
    stratify=y_train
)

clf = LinearSVC()
clf.fit(X_tr, y_tr)

val_pred = clf.predict(X_val)
val_f1 = f1_score(y_val, val_pred, average="macro")
print("Validation Macro F1:", val_f1)

# 3.2 用全部 training data 重新訓練
clf.fit(X_train, y_train)

# 3.3 對 test data 預測
test_pred = clf.predict(X_test)
test_labels = label_encoder.inverse_transform(test_pred)  # 轉回 anger/joy/... 等字串

# 3.4 產生 submission 檔案
submission = pd.DataFrame({
    "id": test_df["id"].values,
    "emotion": test_labels
})

submission.to_csv("submission.csv", index=False)
print("Saved submission.csv")
print(submission.head())
```

```python
### 3. Model Implementation Steps  (Linear SVM + GridSearchCV + submission)

from sklearn.svm import LinearSVC
from sklearn.model_selection import StratifiedKFold, GridSearchCV, train_test_split
from sklearn.metrics import f1_score
import pandas as pd

# 3.1 使用交叉驗證搜尋較好的超參數
param_grid = {
    "C": [0.5, 1.0, 2.0, 4.0],
    "class_weight": [None, "balanced"]
}

base_svm = LinearSVC()
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid = GridSearchCV(
    base_svm,
    param_grid,
    scoring="f1_macro",
    cv=cv,
    n_jobs=-1,
    verbose=1
)

grid.fit(X_train, y_train)
print("Best params:", grid.best_params_)
print("Best CV macro F1:", grid.best_score_)

clf = grid.best_estimator_

# 3.2 再切一個 hold-out validation 看最終效果
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train,
    test_size=0.2,
    random_state=42,
    stratify=y_train
)

clf.fit(X_tr, y_tr)
val_pred = clf.predict(X_val)
val_f1 = f1_score(y_val, val_pred, average="macro")
print("Hold-out Validation Macro F1:", val_f1)

# 3.3 用全部 training data 重訓一次
clf.fit(X_train, y_train)

# 3.4 對 test 做預測 & 產生 submission
test_pred = clf.predict(X_test)
test_labels = label_encoder.inverse_transform(test_pred)

submission = pd.DataFrame({
    "id": test_df["id"].values,
    "emotion": test_labels
})

submission.to_csv("submission_optimized.csv", index=False)
print("Saved submission_optimized.csv")
print(submission.head())
```

```python
### 3. Model Implementation Steps  (Linear SVM + GridSearchCV + 5-fold ensemble + submission)

from sklearn.svm import LinearSVC
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.metrics import f1_score
import numpy as np
import pandas as pd

# 3.1 先用 GridSearchCV 找較好的超參數
param_grid = {
    "C": [0.25, 0.5, 1.0, 2.0, 4.0],
    "class_weight": [None, "balanced"],
}

base_svm = LinearSVC(max_iter=5000)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid = GridSearchCV(
    base_svm,
    param_grid,
    scoring="f1_macro",
    cv=cv,
    n_jobs=-1,
    verbose=1,
)

grid.fit(X_train, y_train)
print("Best params:", grid.best_params_)
print("Best CV macro F1 (GridSearch):", grid.best_score_)

best_params = grid.best_params_

# 3.2 用最佳參數做 5-fold ensemble（每折訓練一個模型，OOF 評估）
svm_models = []
oof_pred = np.zeros_like(y_train)

cv_ens = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)

for fold, (tr_idx, val_idx) in enumerate(cv_ens.split(X_train, y_train)):
    print(f"\n=== Fold {fold+1} / 5 ===")
    model = LinearSVC(
        C=best_params["C"],
        class_weight=best_params["class_weight"],
        max_iter=5000,
    )
    model.fit(X_train[tr_idx], y_train[tr_idx])
    svm_models.append(model)

    val_pred = model.predict(X_train[val_idx])
    oof_pred[val_idx] = val_pred
    fold_f1 = f1_score(y_train[val_idx], val_pred, average="macro")
    print(f"Fold {fold+1} macro F1:", fold_f1)

oof_f1 = f1_score(y_train, oof_pred, average="macro")
print("\nOOF macro F1 from 5-fold ensemble:", oof_f1)

# 3.3 使用 ensemble 對 test 做投票預測
num_classes = len(label_encoder.classes_)
test_votes = np.zeros((X_test.shape[0], num_classes), dtype=int)

for model in svm_models:
    pred = model.predict(X_test)
    for i, p in enumerate(pred):
        test_votes[i, p] += 1

# 多數決
test_pred = test_votes.argmax(axis=1)
test_labels = label_encoder.inverse_transform(test_pred)

# 3.4 產生 submission
submission = pd.DataFrame({
    "id": test_df["id"].values,
    "emotion": test_labels,
})

submission.to_csv("submission_optimized_v2_ensemble.csv", index=False)
print("\nSaved submission_optimized_v2_ensemble.csv")
print(submission.head())
```
